```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/jaybh/OneDrive/Desktop/Machine Learning/Home Price Prediction/dataset/MagicBricks.csv')
```

```python
df1 = df.drop(['Furnishing', 'Status', 'Transaction', 'Type'], axis = 'columns')
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
import math
med = math.floor(df1.Bathroom.median())
df1.Bathroom = df1.Bathroom.fillna(med)
```

```python
df1.isnull().sum()
```

```python
df1.Parking = df1.Parking.fillna(0)
df1['Per_Sqft'] = df1.apply(lambda row: row['Price'] / row['Area'] if pd.isnull(row['Per_Sqft']) else row['Per_Sqft'], axis=1)
```

```python
df1.isnull().sum()
```

```python
df1.head()
```

```python
df1.Locality = df1.Locality.apply(lambda x: x.strip())
location_stats = df1['Locality'].value_counts(ascending=False)
location_df = pd.DataFrame(location_stats)
location_df
```

```python
len(location_stats[location_stats <= 5])
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]
location_stats_less_than_10
df1.Locality = df1.Locality.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
len(df1.Locality.unique())
```

```python
df1.head()
```

```python
df2 = df1[~(df1.Area/df1.BHK < 300)]
df2.shape
```

```python
df2.Per_Sqft.describe()
```

```python
dummies = pd.get_dummies(df2.Locality, dtype=int)
dummies.head(5)
```

```python
df4 = pd.concat([df2, dummies.drop('other', axis = 'columns')], axis = 'columns')
df5 = df4.drop('Locality', axis = 'columns')
df5.head()
```

```python
x = df5.drop('Price', axis = 'columns')
y = df5[['Price']]
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state= 0)
cross_val_score(LinearRegression(), x, y, cv = cv).mean()
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

def find_best_grid(x,y):
    algos = {
        'linear_regression':{
            'model' : LinearRegression(),
            'param' : {
                'fit_intercept' : [True, False],
                'copy_X' : [True, False],
                'positive': [True, False],
            }
        },
        'lasso':{
            'model' : Lasso(),
            'param' : {
                'alpha' : [1,2],
                'max_iter':[1000],
                'selection' : ['random', 'cyclic']
            }
        },
        'decision_tree':{
            'model' : DecisionTreeRegressor(),
            'param' : {
                'criterion' : ["squared_error", "friedman_mse", "absolute_error"],
                'splitter' : ['best', 'random'], 
                'max_features' : [0.5]
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['param'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

d = pd.DataFrame(find_best_grid(x,y))
d
```

```python
cross_val_score(DecisionTreeRegressor(criterion='squared_error', splitter= 'best', max_features = 0.5), x, y ,cv = cv).mean()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

model = DecisionTreeRegressor(criterion='squared_error', splitter= 'best', max_features = 0.5)
X_train, X_test, y_train, y_test = train_test_split(x, y , test_size=0.2, random_state=30)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
import pickle
with open('Delhi_home_prices.pickle','wb') as f:
    pickle.dump(model,f)
```
